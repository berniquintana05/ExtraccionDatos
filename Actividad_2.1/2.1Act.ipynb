```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
#Carga de archivo
df = pd.read_csv("Londres_Inglaterra.csv")
df
```

***Ejercicio A***

```python
#Registros de 5 hosts con nombres diferentes 
filtroa = df.drop_duplicates(subset=['host_name']).iloc[:5]
filtroa.to_csv('HostsDiferentes.csv')
filtroa
```

***Ejercicio B***

```python
#Registros de hosts que hayan unido a Airbnb despuÃ©s del aÃ±o 2020
filtrob=df[df["host_since"] > "31/12/20"]
filtrob.to_csv("HostsDespuÃ©s2020.csv")
filtrob
```

***Ejercicio C***

```python
#Registros de hosts que tengan al menos calificaciÃ³n de 3 en comunicaciÃ³n
filtroc = df[df["review_scores_communication"] >= 3]
filtroc.to_csv("HostCalif3omas.csv")
filtroc
```

***Ejercicio D***

```python
#Registros de filas mÃºltiplos de 200
filtrod = df.iloc[::200]
filtrod.to_csv("FilasMultiplos200.csv")
filtrod
```

***Ejercicio E***

```python
#Registros con tipos de cuarto: "Entire home/apt" y "Private room"
filtroe=df[df["room_type"].isin(["Entire home/apt", "Private room"])]
filtroe.to_csv("TipodeCuarto.csv")
filtroe
```

***Ejercicio F***

```python
#Registros que a lo mucho tengan 4 camas
filtrof=df[df["beds"] <= 4]
filtrof.to_csv("4camasomenos.csv")
filtrof
```

***Ejercicio G***

```python
#Registros que sean superhost
filtrog=df[df["host_is_superhost"] == "t"]
filtrog.to_csv("HostesSuperhost.csv")
filtrog
```
